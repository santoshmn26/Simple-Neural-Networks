```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Magic line to plot in jupyter notebook
%matplotlib inline
```

```python
x = np.arange(0,10)
```

```python
x
```

```python
y=x**2 # x power of 2
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'+')
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,"r--")
plt.xlim(0,6) # start value and end value
plt.ylim(0,50)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("squared")
```
